```python
def run_gradient_descent(learning_rate, w_start=10.0, epochs=50):
    """
    Runs the Gradient Descent algorithm for a simple quadratic function.

    Args:
        learning_rate (float): The learning rate (alpha) for the optimizer.
        w_start (float): The initial value for the weight 'w'.
        epochs (int): The number of iterations to run.

    Returns:
        list: A list containing the history of 'w' values at each epoch.
    """
    w = w_start
    w_history = []
    convergence_count=0

    print(f"\n--- Running Gradient Descent with learning_rate = {learning_rate} ---")

    for i in range(epochs):
        gradient = 2 * w  # The derivative of f(w) = w^2 is 2w
        w = w - learning_rate * gradient  # The GD update rule
        for i in w_history:
          if w==0:
            convergence_count += 1
          if convergence_count >= 5:
              print(f"!!! Early stop at epoch {i+1}: w has been 0 for 5 consecutive steps.")
          break
          w_history.append(w)

        # Optional: Print progress for the first few epochs
        if i < 5 or i > epochs - 6:
            print(f"Epoch {i+1:2d}: w = {w:.6f}")

    print("--------------------------------------------------")
    return w_history
```
